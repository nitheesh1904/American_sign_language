```python
import os
import pandas as pd
from glob import glob


csv_files = ['prev_models\A-F\keypoint.csv', 'prev_models\G-K\keypoint.csv', 'prev_models\L-P\keypoint.csv', 'prev_models/Q-U/keypoint.csv', 'prev_models\V-Z\keypoint.csv']

# Variables to store all data and label offset
all_data = []
temp_label = 0
df_final = pd.DataFrame()
addition = [0, 6, 11, 16, 21]

for i in range(len(csv_files)):
    file = csv_files[i]
    df_temp = pd.read_csv(file, header=None)  # no header assumed
    for idx, cols in df_temp.iterrows():
        cols[0] += addition[i]
        df_final = pd.concat([df_final, cols.to_frame().T], ignore_index=True)
```
